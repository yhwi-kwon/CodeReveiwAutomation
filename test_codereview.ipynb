```python
import pandas as pd
import json
```

```python
# Read the .jsonl file
file_path = 'data/msg-valid.jsonl'

# Load the data from jsonl
with open(file_path, 'r') as file:
    data = [json.loads(line) for line in file]

# Convert to a pandas DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
df['lang'].value_counts()
```

```python
df['y'].value_counts()
```

```python
n = 1000
df_sample = df.sample(n=n, random_state=1115)
df_sample['lang'].value_counts(0)
```

```python
# Write the sampled DataFrame to a new .jsonl file
output_file_path = f'msg-test-{n}.jsonl'
with open(output_file_path, 'w') as file:
    for record in df_sample.to_dict(orient='records'):
        file.write(json.dumps(record) + '\n')
```
