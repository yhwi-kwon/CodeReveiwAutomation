```python
# Import Required Libraries
import pandas as pd

# Load Excel File
file_path = 'data/valid_instances.xlsx'
df = pd.read_excel(file_path, header=1)  # Specify header=0 to use the first row as column names

# Display DataFrame Information
df.info()

# Display First Few Rows
df.head()
```

```python
df['task'].value_counts()
```

```python
df['work'].value_counts()
```

```python
df_clnl = df[df['task'] == 'C2NL']
```

```python
df_clnl['work'].value_counts()
```

```python
df_clnl_4 = df_clnl[df_clnl['work'] == 4]
```

```python
df_clnl_4
```

```python
# Function to extract the first category before '->'
def extract_first_category(labels):
    # Split by newline to handle multiple lines
    lines = labels.split('\n')
    # Extract the first category from each line
    categories = [line.split('->')[0].strip() for line in lines]
    # Join categories with ';'
    return ';'.join(categories)

# Apply function to labels column
df_clnl_4['first_category'] = df_clnl_4['labels'].apply(extract_first_category)

# Display the DataFrame with the new first_category column
df_clnl_4.head()
```

```python
df_clnl_4.to_json('df_clnl_4.jsonl', orient='records', lines=True)
```

```python
df_clnl_4['first_category'].value_counts()
```

```python
df_clnl_4
```

```python
df_target_label = df_clnl_4[['id', 'target', 'labels', 'first_category']]
df_target_label 
```

```python
# Split the 'labels' and 'first_category' columns
df_exploded = df_clnl_4.assign(
    split_labels=df_clnl_4['labels'].str.split('\n'),
    split_categories=df_clnl_4['first_category'].str.split(';')
).explode(['split_labels', 'split_categories'])

# Rename the columns for clarity
df_exploded = df_exploded.rename(columns={'split_labels': 'unique_label', 'split_categories': 'unique_category'})

# Display the resulting DataFrame
df_exploded.head()
```

```python
df_target_label_sorted = df_exploded.sort_values(by='unique_label')
df_target_label_sorted 
```
