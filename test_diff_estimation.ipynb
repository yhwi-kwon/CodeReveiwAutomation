```python
import pandas as pd
import json
```

```python
# Read the .jsonl file
file_path = 'data/cls-test.jsonl'

# Load the data from jsonl
with open(file_path, 'r') as file:
    data = [json.loads(line) for line in file]

# Convert to a pandas DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
df['y'].value_counts()
```

```python
df_sample = df.sample(n=100, random_state=42)
df_sample['lang'].value_counts(0)
```

```python
'''
go	10.262%
java 11.708%
py 16.925%
cpp 9.459%
js 9.859%
rb 4.706%
.cs 3.442%
c 4.616%
php 5.665%
'''
```

```python
# Define the weights for sampling based on 'lang'
weights = {
    'go': 10.262,
    'java': 11.708,
    'py': 16.925,
    'cpp': 9.459,
    'js': 9.859,
    'rb': 4.706,
    '.cs': 3.442,
    'c': 4.616,
    'php': 5.665
}

weights_sum = sum(weights.values())
weights = {k: (v / weights_sum) for k, v in weights.items()}

# Map normalized weights to each row in the DataFrame
df['sampling_weight'] = df['lang'].map(weights)

# Sample 1000 rows based on normalized weights
df_sample = df.sample(n=100, weights='sampling_weight', random_state=42)
df_sample
```

```python
df_sample['lang'].value_counts()
```

```python
# Write the sampled DataFrame to a new .jsonl file
output_file_path = 'sample_output.jsonl'
with open(output_file_path, 'w') as file:
    for record in df_sample.to_dict(orient='records'):
        file.write(json.dumps(record) + '\n')
```
